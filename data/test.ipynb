```python
import csv
import numpy as np
import matplotlib.pyplot as plt

CSV_PATH = "20260414_000226_129/locomotion_dataset.csv"

with open(CSV_PATH, newline="") as f:
    reader = csv.DictReader(f)
    rows = list(reader)

steps = np.array([float(r["sim_step"]) for r in rows], dtype=np.float64)


def _to_float(s):
    s = (s or "").strip()
    if s == "":
        return np.nan
    return float(s)

y = np.array([_to_float(r["lidar_min_distance_xy"]) for r in rows], dtype=np.float64)

plt.figure(figsize=(10, 4))
plt.plot(steps, y, linewidth=0.8)
plt.xlabel("Step (sim_step)")
plt.ylabel("lidar_min_distance (m)")
plt.title("Minimum positive LiDAR range per step")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
