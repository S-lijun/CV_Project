# locomotion_dataset：lidar_min_distance

```python
TRAJECTORY_FOLDER = r"c:/IsaacLab/data/20260409_233507_503"

import pathlib
import pandas as pd
import matplotlib.pyplot as plt

csv_path = pathlib.Path(TRAJECTORY_FOLDER) / "locomotion_dataset.csv"
df = pd.read_csv(csv_path)
y_col = df.columns[-2]
y = pd.to_numeric(df[y_col], errors="coerce")
x = range(len(df))
print(f"rows={len(df)}, y={y_col!r}")

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(list(x), y, lw=0.8)
ax.set_xlabel("step")
ax.set_ylabel(y_col)
ax.set_title(f"min distance to obstacle\n{csv_path}")
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
