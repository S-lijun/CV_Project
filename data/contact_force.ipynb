```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("contact_force.csv")

# step
steps = df["step"]


links = set()
for col in df.columns:
    if col != "step":
        links.add(col[:-2])  

links = sorted(list(links))

print("Detected links:", links)


def plot_axis(axis):
    plt.figure(figsize=(10,5))

    for link in links:
        col = f"{link}_{axis}"
        if col in df.columns:
            plt.plot(steps, df[col], label=link)

    plt.xlabel("Step")
    plt.ylabel(f"Force ({axis})")
    plt.title(f"Contact Force - {axis.upper()} axis")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') 
    plt.grid()
    plt.tight_layout()
    plt.show()


plot_axis("x")
plot_axis("y")
plot_axis("z")
```
